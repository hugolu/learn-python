```python
#-*- coding: utf-8 -*-
import matplotlib.pyplot as plt
import numpy as np

points = np.linspace(-1, 1, 201)
xs, ys = np.meshgrid(points, points)
zs = np.sqrt(xs ** 2 +  ys ** 2)

plt.imshow(zs, cmap=plt.cm.gray)
plt.colorbar()

plt.title('Image plot of $\sqrt{x^2 + y^2}$ for a grid of values')
plt.xlabel(u'x轴')
plt.ylabel(u'y轴')

plt.show()
```
