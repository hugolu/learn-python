# 风格管理
## 控制图片外观
http://seaborn.pydata.org/tutorial/aesthetics.html

```python
%pylab inline
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
np.random.seed(sum(map(ord, "aesthetics")))
```

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
# 预设用 matplotlib 画出来的图形
sinplot()
```

```python
# 使用 seaborn 画出来的图形
import seaborn as sns
sinplot()
```

seaborn 将 matplotlib 参数分成两部分。
- 第一：图形美学风格 - axes_style()、set_style()
- 第二：缩放图片元素 - plotting_context()、set_context()

### 使用 `axes_style()` 与 `set_style()` 设定图片风格

五种主题：darkgrid（预设）, whitegrid, dark, white, and ticks

```python
sns.set_style("whitegrid")
sinplot()
```

```python
sns.set_style("dark")
sinplot()
```

```python
sns.set_style("white")
sinplot()
```

```python
sns.set_style("ticks")
sinplot()
```

### 使用 `despine()` 移除外框

```python
sinplot()
sns.despine()
```

```python
sinplot()
sns.set_style('white')
sns.despine(offset=10, trim=True) # 位移
```

```python
sinplot()
sns.set_style('white')
sns.despine(left=True) # 移除边框
```

### 暂时设定图片风格

使用 `with sns.axes_style():` 的语法暂时设定图片参数 

```python
with sns.axes_style("darkgrid"):
    plt.subplot(211)
    sinplot()
plt.subplot(212)
sinplot(-1)
```

### 复写 Seaborn 风格元素

透过传递参数给 `axes_style()` 或 `set_style()` 复写图片风格

```python
# 不传递参数，查看风格设定
sns.axes_style()
```

```python
sns.set_style("darkgrid", {"axes.facecolor": ".2"})
sinplot()
```

### 使用 `set()` 设定全域参数

```python
# 修改图片大小
sns.set(rc={"figure.figsize": (6, 6)})
```

```python
sinplot()
```

```python
# 重置全域参数
sns.set()
```

```python
sinplot()
```

### 使用 `plotting_context()` 与 `set_context()` 调整图片元素

四种预设情景(图片大小相关)：`paper`, `notebook`, `talk`, `poster`。预设为 `notebook`

```python
# 重置设定
sns.set()
```

```python
sns.set_context("talk")
#plt.figure(figsize=(8, 6))
sinplot()
```

```python
sns.set_context("poster")
#plt.figure(figsize=(8, 6))
sinplot()
```

```python
# 透过 `rc` 复写图形参数
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 5})
sinplot()
```

```python
# 重置
sns.set()
```

#### 使用 `cubehelix_palette()` 的顺序调色板

```python
sns.palplot(sns.color_palette("cubehelix", 8))
```

```python
# cubehelix 调色板产生在亮度、彩度线性递增、递减的效果
sns.palplot(sns.cubehelix_palette(8))
```

```python
# start: 0~3 
# rot(rotation): -1~1
sns.palplot(sns.cubehelix_palette(8, start=.5, rot=-.75))
```

```python
sns.palplot(sns.cubehelix_palette(8, start=2, rot=0, dark=0, light=.95, reverse=True))
```

```python
x, y = np.random.multivariate_normal([0, 0], [[1, -.5], [-.5, 1]], size=300).T
cmap = sns.cubehelix_palette(light=1, as_cmap=True) # 返回当成 colormap object 的调色板
sns.kdeplot(x, y, cmap=cmap, shade=True);
```
