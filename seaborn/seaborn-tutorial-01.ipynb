## Controlling figure aesthetics

```python
%pylab inline
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
np.random.seed(sum(map(ord, "aesthetics")))
```

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
# 预设用 matplotlib 画出来的图形
sinplot()
```

```python
# 使用 seaborn 画出来的图形
import seaborn as sns
sinplot()
```

seaborn 将 matplotlib 参数分成两部分。
- 第一：图形美学风格 - axes_style()、set_style()
- 第二：缩放图片元素 - plotting_context()、set_context()

### 使用 `axes_style()` 与 `set_style()` 设定图片风格

五种主题：darkgrid（预设）, whitegrid, dark, white, and ticks

```python
sns.set_style("whitegrid")
sinplot()
```

```python
sns.set_style("dark")
sinplot()
```

```python
sns.set_style("white")
sinplot()
```

```python
sns.set_style("ticks")
sinplot()
```

### 使用 `despine()` 移除外框

```python
sinplot()
sns.despine()
```

```python
sinplot()
sns.set_style('white')
sns.despine(offset=10, trim=True) # 位移
```

```python
sinplot()
sns.set_style('white')
sns.despine(left=True) # 移除边框
```
