```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
iris = sns.load_dataset("iris")
tips = sns.load_dataset('tips')
```

## set_style( ), set( )

```python
sns.set_style("whitegrid")
plt.plot(np.arange(10))
plt.show()
```

```python
# 设置主题
sns.set(style="white", palette="muted", color_codes=True)

plt.plot(np.arange(10))
plt.show()
```

## distplot( )  kdeplot( )

```python
fig, axes = plt.subplots(1,2)  
sns.distplot(iris['petal_length'], ax = axes[0], kde = True, rug = True)        # kde 密度曲线  rug 边际毛毯  
sns.kdeplot(iris['petal_length'], ax = axes[1], shade=True)                     # shade  阴影                         
plt.show()
```

```python
sns.set(palette="muted", color_codes=True)

rs = np.random.RandomState(10)
d = rs.normal(size=100)
f, axes = plt.subplots(2, 2, figsize=(7, 7), sharex=True)

sns.distplot(d, kde=False, color="b", ax=axes[0, 0])
sns.distplot(d, hist=False, rug=True, color="r", ax=axes[0, 1])
sns.distplot(d, hist=False, color="g", kde_kws={"shade": True}, ax=axes[1, 0])
sns.distplot(d, color="m", ax=axes[1, 1])
plt.show()
```

## boxplot( )

```python
sns.boxplot(x = iris['species'], y = iris['sepal_width'])  
plt.show()  
```

```python
sns.jointplot("total_bill", "tip", tips)
plt.show()
```

```python
sns.jointplot("total_bill", "tip", tips, kind='reg')
plt.show()
```

##  pairplot( )

```python
sns.set() #使用默认配色  
sns.pairplot(iris,hue="species") #hue 选择分类列  
plt.show() 
```

```python
sns.pairplot(iris, vars=["sepal_width", "sepal_length"], hue='species', palette="husl")    
plt.show()
```

## FacetGrid( )

```python
g = sns.FacetGrid(tips, col="time",  row="smoker")
g = g.map(plt.hist, "total_bill",  color="r")
plt.show()
```
