```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

## set_style( ), set( )

```python
sns.set_style("whitegrid")
plt.plot(np.arange(10))
plt.show()
```

```python
# 设置主题
sns.set(style="white", palette="muted", color_codes=True)

plt.plot(np.arange(10))
plt.show()
```

## distplot( )  kdeplot( )

```python
from sklearn import datasets
iris = datasets.load_iris()
df_data = DataFrame(iris.data, columns=['sepal length', 'sepal width', 'petal length', 'petal width'])
df_target = DataFrame(iris.target, columns=['class'])
df_iris = pd.merge(df_data, df_target, right_index=True, left_index=True)
```

```python
fig, axes = plt.subplots(1,2)  
sns.distplot(df_iris['petal length'], ax = axes[0], kde = True, rug = True)        # kde 密度曲线  rug 边际毛毯  
sns.kdeplot(df_iris['petal length'], ax = axes[1], shade=True)                     # shade  阴影                         
plt.show()
```

```python
sns.set( palette="muted", color_codes=True)

rs = np.random.RandomState(10)
d = rs.normal(size=100)
f, axes = plt.subplots(2, 2, figsize=(7, 7), sharex=True)

sns.distplot(d, kde=False, color="b", ax=axes[0, 0])
sns.distplot(d, hist=False, rug=True, color="r", ax=axes[0, 1])
sns.distplot(d, hist=False, color="g", kde_kws={"shade": True}, ax=axes[1, 0])
sns.distplot(d, color="m", ax=axes[1, 1])
plt.show()
```

## boxplot( )

```python
sns.boxplot(x = df_iris['class'],y = df_iris['sepal width'])  
plt.show()  
```

```python
tips = pd.read_csv('tips.csv')
sns.jointplot("total_bill", "tip", tips)
plt.show()
```

```python
sns.jointplot("total_bill", "tip", tips, kind='reg')
plt.show()
```
