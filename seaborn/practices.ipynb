```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

import seaborn as sns
```

## 离散数据可视化

### 单变数分布图

```python
x = np.random.normal(size=100)
```

```python
sns.distplot(x)
sns.rugplot(x)
```

### 双变数分布图

```python
mean, cov = [0, 1], [(1, .5), (.5, 1)]
data = np.random.multivariate_normal(mean, cov, 200)
df = pd.DataFrame(data, columns=["x", "y"])
```

```python
sns.jointplot(x="x", y="y", data=df)
```

```python
sns.jointplot(x="x", y="y", data=df, kind='hex')
```

```python
sns.jointplot(x="x", y="y", data=df, kind='kde')
```

### 数据两两关系

```python
iris = sns.load_dataset('iris')
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue='species', diag_kind='kde')
```

## 线性关系可视化

```python
tips = sns.load_dataset("tips")
```

```python
sns.regplot(x='total_bill', y='tip', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', hue='smoker', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', hue='smoker', col='time', data=tips)
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", col="time", row="sex", data=tips)
```

```python
sns.pairplot(tips, x_vars=["total_bill", "size"], y_vars=["tip"], hue="smoker", size=5, aspect=.8, kind="reg");
```

## 分类数据可视化

```python
sns.stripplot(x="day", y="total_bill", data=tips)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
sns.swarmplot(x='day', y='total_bill', hue='sex', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', hue='sex', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', hue='sex', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', hue='sex', data=tips, split=True)
```

### 类别统计估计

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
sns.barplot(x='sex', y='survived', hue='class', data=titanic)
```

```python
sns.countplot(x="deck", data=titanic, palette="Greens_d")
```

```python
sns.countplot(x='deck',  hue='class', data=titanic)
```

### 多面类别图

```python
sns.factorplot(x="day", y="total_bill", hue="smoker", data=tips, kind="bar")
```

```python
sns.factorplot(x="day", y="total_bill", hue="smoker", col='time', data=tips, kind="bar")
```
